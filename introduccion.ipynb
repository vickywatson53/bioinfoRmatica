## ‚òù Operador de entubamiento (*Pipe*)

|Operador|Significado|
|:------:|---------|
|\|>| toma la salida de la operaci√≥n de la izquierda y la utiliza como la entrada de la operaci√≥n de la derecha |

**Nota:** Antes de R 4.1 no funciona, pero existe el operador `%>%` que √∫nicamente sirve para determinadas funciones pero a√∫n es muy frecuente encontrarlo en diferentes p√°ginas y documentaci√≥n, se debe preferir el mas nuevo `|>`


## üìñ Anidaci√≥n vs entubamiento

En la anidaci√≥n la funci√≥n mas "adentro" se ejecuta primero:



```
tercer_funci√≥n(segunda_funcion(primer_funcion(datos)))
```

En el entubamiento la ejecuci√≥n es de izquierda a derecha



```
datos |> primer_funcion() |> segunda_funcion() |> tercer_funcion()
```
