## ☝ Listas

Las listas pueden guardar diferentes tipos de objetos a diferencia de los vectores

un objeto iniciado así intentará convertir los datos a un sólo tipo

```
vec <- c("Tengo", 1 , "torta")
```

En cambio cómo una lista pueden convivir objetos de distinto tipo sin que estos cambien de clase

```
lst <- list("Tengo", 1 ,"torta")
```

Las listas incluso pueden almacenar objetos mas complejos como vectores o matrices)

```
lst2 <- list(califs, alumnos)

```

Una característica que suele ser útil es asignarles nombres a los  elementos de una lista

```
lst2 <- list(calificaciones = califs ,estudiantes = alumnos)
```

De ese modo si quiero conocer, por ejemplo, los nombres de los estudiantes puedo usar el comando `print(lst2[2])` o bién el comando `print(lst2["estudiantes"]`

Para conocer los nombres asignados a un objeto podemos usar el comando `names()`

## ☝ Matrices

Las matrices son estructuras de datos que almacenan **datos númericos** en un formato de tabla, las columnas de la tabla deben tener el mismo número de elementos.

```
genes <- c("p53", "egfr", "dubr", "linc00883","gata")
l2fc <- c(1.2,3.1,-8,0.01,-0.2)
padj <- c(0.001,.1,0.00003,1,1)

mtx <- cbind(log2fc = l2fc, pVal = padj)
rownames(mtx)<- genes
```

````r
genes <- c("p53", "egfr", "dubr", "linc00883","gata")
l2fc <- c(1.2,3.1,-8,0.01,-0.2)
padj <- c(0.001,.1,0.00003,1,1)

mtx <- cbind(log2fc = l2fc, pVal = padj)
rownames(mtx)<- genes

class(mtx)
typeof(mtx)

mtx
````

## ☝ Data Frames

Los data frames son como hojas de Excel, pueden incluir diferentes clases de datos pero no es recomendable incluir vectores de mas de un elemento dentro de un Data Frame.

```
genes <- c("p53", "egfr", "dubr", "linc00883","gata")
l2fc <- c(1.2,3.1,-8,0.01,-0.2)
padj <- c(0.001,.1,0.00003,1,1)

df <- data.frame(geneName = genes, log2fc = l2fc, pVal = padj)
```

````r
genes <- c("p53", "egfr", "dubr", "linc00883","gata")
l2fc <- c(1.2,3.1,-8,0.01,-0.2)
padj <- c(0.001,.1,0.00003,1,1)

df <- data.frame(geneName = genes, log2fc = l2fc, pVal = padj)

class(df)
typeof(df)

df
````

## ☝ Accediendo a los datos de matrices o data frames

Para ver o trabajar sobre elementos de una matriz o data frame usamos los corchetes cuadrados tal como con las listas o vectores, pero consideramos que tienen dos dimensiones, por tanto:

Para conocer la primera fila de una matriz puedo hacer: (`mtx` representa el nombre de una matriz cualquiera, debes sustituirlo por el nombre de tu matriz)

```
mtx[1,]
```

y la primer columna:

```
mtx[,1]
```

el primer elemento de la segunda fila:

```
mtx[2,1]
```

En el caso de los data frames podemos seguir el mismo método pero también podemos directamente hacer referencia a sus columnas


Esto nos devuelve el contenido de una columna denominada `geneName` en un data frame llamado `df`

```
df$geneName
```
por lo que el tercer elemento de esa columna lo podemos conocer así

```
df$geneName[3]
```
e inclusive **sustituirlo** así:

```
df$geneName[3] <- "CTCF"
```
